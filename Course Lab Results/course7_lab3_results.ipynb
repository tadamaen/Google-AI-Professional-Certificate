### **Step 1: Define your app prototype**

#### **Prompt:**
I want to make an interactive dashboard that allows me to visualize data in many different ways. I am going to provide the data for you, so please set up the dashboard with it accurately.

#### **Google AI Studio Result:**

I have built a highly interactive, responsive, and visually gorgeous Interactive Data Dashboard designed to host and analyze the exact 116 clothing transaction records you provided. Every calculation, statistic, and chart updates instantly in real-time.

Here is a summary of the accomplishments and design assets created:

**🎨 Visual & Layout Design**

- Space Grotesk & Inter Pairing: Integrated clean, modern display typography for dashboard headers and metric widgets, balanced with highly legible sans-serif elements for transaction lines.
- Bento-Grid Metric Deck: Configured 5 premium key KPI banners showcasing live, dynamic updates for Gross Revenue, Checkout Counts, Average Order Value (AOV), Peak Selling Product, and Timeline Span.
- Intuitive Tabs Navigation: Designed fluid animated transitions (using motion) to quickly pivot between different views.
  
**📊 Dashboard Analytical Views (Visualizing in Many Ways)**

1. Sales Trend & Distribution (Overview Tab):
- Revenue Stream: Displays sales trajectories over time, with instant toggles between Daily and Weekly groupings as well as different chart types (Area, Line, Bar).
- Payment Channels split: Visualizes checkout gateway distributions in a premium centered Donut Pie Chart representation.
- Top Products popularity: Shows gross billing share for the catalog's leading performers using a horizontal Bar Chart with distinct color-category blocks.

2. Product Metrics (Product Deep-Dive Tab):
- Lineup Compare Matchup: Allows you to multi-select checkbox tags for any product (or all 13 products) and see their comparative sales curves mapped together over the timeline.
- Catalog Ledger: An in-depth table measuring total gross revenues, precise unit sale units, average pricing models, and dominant payment methods for each individual cloth type.

3. Financial Deep-Dive (Financial Channels Tab):
- Gateway Worth: Compiles and diagrams Average Order Values (AOV) across digital gateways, contrasting which payment methods capture the highest cart sizes.
- Cumulative Timeline: Maps stacked areas representing billing modes day-by-day to track shifting checkout preferences.

4. Temporal Triggers (Timing Tab):
- Sales by Day of the Week: Identifies peak sales periods split across Monday through Sunday.
- Progression Timeline: Maps cumulative monthly progression to monitor performance month-over-month.

5. Order Book & File Engine (Order Booklet & CSV Tab):
- Live Filtration: Search by product name or order number, select specific payment rules, slide price thresholds (min/max bounds), or restrict dates.
- Inline Order Addition Form: Add custom sales records directly on the screen to watch every single chart on all tabs instantly update in real-time.
- Dynamic Data Import (Drag-and-Drop / Browse): Drag and drop any other sales CSV sheet directly onto the page to immediately parse, map, and visualize entirely different transaction sets.
- Clean CSV Downloader: Export your filtered, updated, or custom transactions list back as a clean CSV file with a single click.
