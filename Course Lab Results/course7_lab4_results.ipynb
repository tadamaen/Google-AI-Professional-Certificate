### **Step 1: Describe the app you want to build**

#### **Prompt:**
I want to build an app that takes customer support emails and does sentiment analysis, topic classification, and issue extraction to create a prioritized dashboard of customer complaints and requests that helps me respond faster to urgent issues and improve customer satisfaction.

#### **Google AI Studio Result:**
I have designed and implemented a full-stack, AI-powered Customer Support Email Analyzer dashboard. The application is built using a React + Vite frontend, an Express backend, and leverages the official server-side @google/genai SDK to classify issues in real-time without exposing sensitive API credentials.

Here is a summary of the accomplishments and core systems designed for this platform:

**Key Visual & Functional Implementations**

1. Analytical Performance HUD:
- Critical Ticket Tracker: Automatically counts and highlights unresolved Urgent and High priority issues that require immediate action to preserve service level agreements (SLAs).
- Avg Customer Sentiment: Aggregates tone scores inside your inbox so you can monitor changes in overall satisfaction.
- Resolution Success Rate: Displays a visual indicator of solved vs. pending cases.

2. Refined Multi-Filter Support Inbox:
- Includes real-text search querying.
- Allows instant segmentation by Urgency level (Urgent, High, Medium, Low), Topic Category (Billing & Payments, Technical Bug, Login & Accounts, Feature Request, General Inquiry), Resolution Status (New, Investigating, Replied, Resolved), and Customer Sentiment Mood (Angry, Frustrated, Confused, Neutral, Pleased).
- Allows support supervisors to manual-override any classification or priority level to adjust routing.

3. Gemini AI Deep-Extraction Engine:
- Parses incoming text to output a concise 1-sentence recap.
- Extracts specific technical complaints, error messages, and failed operations.
- Outputs customized checklist rows so agents can mark off resolved issues as they work through a ticket.
- Identifies emotional factors to explain customer mood.

4. Response Copilot Drafting Chamber:
- Generates contextual email responses based on the email's content and extracted issues using Gemini-3.5-flash.
- Includes selectable Tone Presets (Apologetic & Sorrowful, Professional & Balanced, Technical & Instructional, Direct & Fast) to match any custom brand communication standard.
- Fully interactive raw textbox editor so you can tweak drafts before copying with single-click feedback.

5. Local Persistent JSON Database:
- Uses local server-side files to store support tickets, draft changes, overrides, and notes. Your updates remain persistent across server recycles.
- Auto-Seeded: The application loads with five highly detailed support email scenarios pre-analyzed by Gemini. You can also click the Reset Database button in the header at any time to recover the original seeds.

6. Interactive Live Email Simulator:
- Allows you to paste your own custom emails or instantly load our built-in simulator presets (e.g., Hostile Cancellation, Security Audits, iPad Feature Suggestions) with active progress logging to visualize the analysis steps.

7. Aesthetic Typography & Visualization:
- Uses Space Grotesk display typography for geometric headers, Inter for clean readability, and JetBrains Mono for serial ID tracking and status metrics.
- Embeds responsive Recharts charts to map emotional volume and topic distributions.
