### **Step 1: Brainstorm new product ideas**

#### **Prompt:**
You are a food advertiser in the food and beverage industry. My company is developing a social media advertisement on the opening of Kim Kee Chicken Rice in Balestier. Help me brainstorm possible advertising strategies that I could adopt in order to help Kim Kee Chicken Rice improve their business and attract a wide base of customers.

#### **Gemini Result:**

Balestier Road is essentially the ground zero of chicken rice in Singapore. Opening Kim Kee there means going head-to-head with heritage giants like Loy Kee and Boon Tong Kee. To attract a wide base of customers, your social media strategy can't just be "we have good food" — it needs to cut through the noise with highly shareable, targeted angles.Here is a breakdown of how we can position Kim Kee Chicken Rice to stand out in a saturated market.

**1. The "Auntie Network" (Dark Social) Strategy:**

Don't just rely on standard Instagram posts or Facebook ads. A massive driver of local food traffic comes from older generations who love to forward "good deals" and "must-try" lists in their family chat groups.

The Format: Create promotional content that looks and reads like a bite-sized local news update or a WhatsApp-friendly listicle (e.g., "3 Reasons Why Balestier's Newest Chicken Rice is Selling Out by 2 PM").

The Hook: Emphasize keywords they care about: "Fresh Kampung Chicken," "Zero MSG," or "Opening Promo for Seniors."

The Goal: Get parents to forward these news-style graphics and articles to their kids and extended families, effectively doing your distribution for you.

**2. The "Caught on Camera" Teaser:**

Lean into the way people organically consume localized content. We know that mildly dramatic, "caught on camera" style videos regarding local occurrences get shared rapidly.

The Format: Produce a viral-style, slightly shaky-cam video of an "incident"—like a customer obsessively guarding their last piece of roasted chicken at the Balestier outlet, or a faux-dramatic standoff between two people over the last bottle of house-made chili.

The Hook: It mimics the raw, shareable social media content that friends constantly send each other. 

The Goal: Drive curiosity and shares before people even realize it's a clever ad for a new opening.

**3. Position Against the Giants:**

You need to clearly communicate why someone should walk past Boon Tong Kee to eat at Kim Kee. We can use targeted social media ads to highlight your specific competitive advantage. Use the tool below to visualize where Kim Kee fits in the Balestier market landscape.

Key insight: If the heritage brands own the "Traditional/Premium" quadrant, Kim Kee's ads should aggressively target the whitespace—perhaps marketing as "Modern/Affordable" or offering late-night supper deals that the older restaurants don't cater to.

**4. Hyper-Local Commuter Targeting:**

Since Balestier is a major thoroughfare but lacks an immediate MRT station, accessibility is an objection you have to overcome.

The Format: Geo-targeted ads served to people during their evening commute (e.g., those transiting through nearby hubs like Novena or Toa Payoh).

The Hook: "Skip the cooking tonight. Pre-order Kim Kee on your commute and pick up curbside on Balestier."

The Goal: Turn the friction of getting to Balestier into a convenience play for busy professionals heading home.

### **Step 3: Research industry trends**

#### **Prompt:**
What are some current industry trends around food advertising in Singapore? I'd like to factor those insights into my decision.

#### **Gemini Result:**

To make the best decision for Kim Kee, you want to align with how Singaporeans are actually discovering food right now. The local digital landscape has shifted dramatically, moving away from hyper-curated "food porn" toward raw, community-driven discovery.

Here are the critical food advertising trends shaping the Singapore market that you should factor into your strategy.

**1. Unpolished Authenticity & "Micro-Dramas"**

The era of perfectly lit, overly staged studio photography of a chicken rice platter is losing its grip. Singaporean diners are experiencing "aesthetic fatigue."

The Reality: Audiences actively crave raw, human-centric, and slightly chaotic content. Content that feels like it was filmed on a phone by a real customer or a back-of-house staff member vastly outperforms boardroom-designed ads.

Application for Kim Kee: This strongly validates the Caught on Camera idea. A slightly messy, fast-paced video showing the intense rush of the kitchen, the rapid rhythmic chopping of the chicken, or a lighthearted customer interaction will feel native to a user's feed, driving higher engagement than a glossy commercial.

**2. The "TikTok to Table" Search Funnel**

Social media is officially the new search engine for food. When looking for a place to eat in a specific neighborhood, consumers are bypassing traditional search engines and typing queries directly into TikTok, Instagram Reels, and increasingly, Xiaohongshu (Little Red Book).

The Reality: F&B discovery heavily relies on short-form vertical video. Diners want to experience the food contextually before they visit—they want to see the texture of the rice, the sheen on the roasted chicken skin, and the exact viscosity of the chili sauce.

Application for Kim Kee: Your content must be optimized for social SEO. Every piece of short-form video needs clear, keyword-rich captions, accurate location tags, and on-screen text so it shows up when someone searches for food options near Novena, Toa Payoh, or Balestier.

**3. Hyper-Local Interception (The 2km Radius)**

With customer acquisition costs rising across major ad platforms, mass-targeting the whole island for a single-outlet launch is incredibly inefficient.

The Reality: Smart F&B brands are leveraging hyper-targeted, geo-fenced paid ads. They drop a tight virtual pin directly around their shop and only serve ads to users within a 2-kilometer radius during high-intent time blocks (like 11:30 AM to 1:30 PM for lunch, and 5:30 PM to 8:30 PM for dinner and supper).

Application for Kim Kee: This gives a massive boost to the Hyper-Local Commuter Targeting strategy. Instead of hoping someone from Jurong makes the trek, focus your ad spend entirely on the dense residential and office crowds in the immediate Balestier, Whampoa, and Toa Payoh areas right when they are deciding what to eat.

**4. The Rise of "Dark Social" and Micro-Communities**

While public feeds drive awareness, the actual decision to try a place often happens in private messaging spaces like WhatsApp and Telegram.

The Reality: Peer-to-peer recommendations carry the highest trust conversion rate in Singapore. Word-of-mouth has digitized into family group chats and neighborhood Telegram channels where food deals and new openings are shared constantly.

Application for Kim Kee: This highlights the potential of the Auntie Network strategy. Making your promotions easily downloadable, shareable, and formatted specifically to look good when forwarded on WhatsApp can unlock a massive, highly loyal demographic that traditional social ads completely miss.

The Takeaway: If you want a strategy that hits the ground running, combining Hyper-Local Interception (for immediate, predictable foot traffic) with Short-Form Video Search Content (to build long-term local discovery) will give Kim Kee the highest return on your ad spend.
