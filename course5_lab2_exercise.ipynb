## **Build with AI: Presentation slides**

### **Lab Tasks**

#### **Step 1: Build the slide deck**

Click the “New chat” button to start a new chat with Gemini. Upload the document that you want to turn into a slide deck. Click the “Plus” icon in the Gemini prompt box, where you can either upload a file, or you can add it from Drive. 

You can use your own document or use a sample document outlining a service overview from an IT business. Then, click the Tools bar and select Canvas.

Paste the prompt below into the input bar, editing the bracketed portions to match your needs:

Acting as an expert [type of expertise you need, ex: designer] help me create a text-only slide deck based on [document you have supplied, ex: new refund policy].

Always double check Gemini’s work. Make sure that the information in the slides matches the information in the doc you gave it to ensure AI hasn’t hallucinated or missed anything.

If you notice Gemini hallucinated or missed a key detail from your uploaded document, you can correct Gemini with the following prompt: 

You missed [specific detail] from the attached source document, please add it to [slide number].

#### **Step 2: Edit your slides**

You can ask Gemini to make changes to the deck as well, such as adding information or changing the color of the slides, for example:

Update the last slide to say: [edit you want Gemini to make].

Or: Make the tone of the entire deck more enthusiastic and sales-focused.

Or: Add a column to the table on Slide 5 that includes [specific data].

Make changes as you need them. 

#### **Step 3: Customize for different audiences**

You can use Gemini to adjust the slides for a different audience. Give Gemini details about what would make the slides better for that particular audience. 

Paste the prompt below:

Change the slide deck to be for [new type deck you need, ex: an internal onboarding deck for new hires]. Use only the information provided in the slides.

Again, make sure all the information is correct, and ask Gemini to make any changes you need. 

Once you get it to a place where you are happy with the slides, you can export it to Google Slides, where you can share it with your team and continue editing the slides in a more hands-on manner. You can also download the slides as a PDF.
