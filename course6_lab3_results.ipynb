### **Step 3: Use your data to build a narrative**

#### **Prompt:**
You are a seasoned data analyst. For an e-commerce site, why might the average revenue per session vary significantly across different channels and device categories, with desktop users generally generating more revenue per session than mobile users?

#### **Gemini Result:**

As a data analyst, when I see significant variance in Average Revenue Per Session (ARPS) across channels and devices, I always start by breaking down the metric into its two core components:

ARPS = Ecommerce Conversion Rate × Average Order Value (AOV)

If desktop is driving higher revenue per session than mobile, it means desktop users are either converting more frequently, buying more expensive items, or both. Here is why this split happens and why channel matters just as much.

What to look for: Notice how a massive volume of traffic flows through mobile, but the "Purchase" outcome is disproportionately fueled by desktop traffic.

Even though mobile generates the lion's share of overall traffic for most modern e-commerce sites, desktop consistently wins on ARPS. This comes down to user psychology and user experience.

**1. The "Research vs. Buy" Divide**

Mobile devices are top-of-funnel discovery tools. Users browse on mobile while commuting, watching TV, or waiting in line. They are window shopping. When it’s time to actually pull out a credit card—especially for a complex or high-ticket item—users often switch to a desktop. Desktop represents a "sit-down and focus" mindset.

**2. Friction in the Funnel**

- Checkout flows are inherently more difficult on a 6-inch screen.
- Navigation: Comparing three different products is easy on a desktop with multiple browser tabs. On mobile, it requires endless back-and-forth tapping, which causes fatigue and abandonment.
- Data Entry: Typing shipping addresses and credit card numbers is cumbersome on mobile (though digital wallets like Apple Pay are shrinking this gap).

**3. Trust and Ticket Size**

For low-cost impulse buys (like a `$20` t-shirt), mobile converts fine. But as the Average Order Value goes up (e.g., a `$1,500` laptop or custom furniture), consumers subconsciously prefer the perceived security and comprehensive view of a desktop monitor. They want to read the fine print, view high-res images, and check return policies before dropping serious cash.

**4. The Compounding Effect**

These variables don't act in isolation - they compound. A session originating from a TikTok ad on a smartphone is the perfect storm for a low ARPS: low-intent scrolling combined with a high-friction device. Conversely, a session originating from a promotional email opened on a desktop yields a massive ARPS: high trust combined with a low-friction purchasing environment.
